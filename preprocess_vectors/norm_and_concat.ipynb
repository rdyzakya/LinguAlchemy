# Norm

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
ppl = pd.read_json("./ppl_vectors/ppl_vectors.json")
```

```python
ppl.head()
```

```python
ppl = ppl.transpose()
```

```python
ppl.head()
```

```python
ppl_std = pd.DataFrame(StandardScaler().fit_transform(ppl), index=ppl.index, columns=ppl.columns)
ppl_minmax = pd.DataFrame(MinMaxScaler().fit_transform(ppl), index=ppl.index, columns=ppl.columns)

ppl_std = ppl_std.transpose().to_dict()
ppl_minmax = ppl_minmax.transpose().to_dict()

for k, v in ppl_std.items():
    ppl_std[k] = list(v.values())

for k, v in ppl_minmax.items():
    ppl_minmax[k] = list(v.values())
```

```python
import json

with open("./ppl_vectors/ppl_vectors_std_scale.json", 'w') as fp:
    json.dump(ppl_std, fp)

with open("./ppl_vectors/ppl_vectors_minmax_scale.json", 'w') as fp:
    json.dump(ppl_minmax, fp)
```

# Concat

```python
import os

os.makedirs("./uriel_ppl_vectors", exist_ok=True)

for fname in os.listdir("./uriel_vectors"):
    path = os.path.join("./uriel_vectors", fname)
    with open(path, 'r') as fp:
        vector = json.load(fp)
    res_std = {}
    for k in ppl_std.keys():
        if k not in vector.keys():
            res_std[k] = ["--" for _ in range(len(list(vector.values())[0]))] + ppl_std[k]
        else:
            res_std[k] = vector[k] + ppl_std[k]
    res_minmax = {}
    for k in ppl_std.keys():
        if k not in vector.keys():
            res_minmax[k] = ["--" for _ in range(len(list(vector.values())[0]))] + ppl_minmax[k]
        else:
            res_minmax[k] = vector[k] + ppl_minmax[k]
    with open(os.path.join("./uriel_ppl_vectors", fname.replace(".json", "_std_scaled.json")), 'w') as fp:
        json.dump(res_std, fp)
    with open(os.path.join("./uriel_ppl_vectors", fname.replace(".json", "_minmax_scaled.json")), 'w') as fp:
        json.dump(res_minmax, fp)
```
